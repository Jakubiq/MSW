# Monte Carlo model
- vybral jsem si monte carlo model protože jsem našel zajímavý způsob hledání přibližné hodnoty $\pi$ (pí), pomocí vrhání bodů do čtverce a následné sledování kolik z těchto bodů zkončí v kruhu jenž je vepsaný do čtverce. 

```python
import numpy as np
import matplotlib.pyplot as plt

# čím více bodů vygenerujem tím přesnější bude výsledek
N = 10000

x = np.random.uniform(-1, 1, N)
y = np.random.uniform(-1, 1, N)

inside = x**2 + y**2 < 1

pi_estimate = 4 * np.sum(inside) / N

print(f"Odhad hodnoty π: {pi_estimate}")

plt.figure(figsize=(6,6))
plt.scatter(x[inside], y[inside], color='blue', s=1, label='Uvnitř kruhu')
plt.scatter(x[~inside], y[~inside], color='red', s=1, label='Mimo kruh')
circle = plt.Circle((0, 0), 1, color='black', fill=False)
plt.gca().add_artist(circle)
plt.title('Monte Carlo odhad čísla π')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.legend()
plt.show()
```

# Házení kostkou



```python
import random

def simulate_dice_roll(num_rolls):
    results = []
    for _ in range(num_rolls):
        # Hod první kostkou 
        die1 = random.randint(1, 6)
        # Hod druhou kostko
        die2 = random.randint(1, 6)
        results.append(die1 + die2)
    return results

def monte_carlo_dice_probability(num_simulations, target_sum):
    # Spustíme simulace
    all_sums = simulate_dice_roll(num_simulations)

    count_target_sum = 0
    for s in all_sums:
        if s == target_sum:
            count_target_sum += 1

    # Vypočítáme odhadovanou pravděpodobnost
    estimated_probability = count_target_sum / num_simulations
    return estimated_probability

# --- simulace ---
num_simulations = 1000000  
target_sum_to_find = 7

print(f"Spouštím Monte Carlo simulaci pro {num_simulations} hodů kostkou.")

estimated_prob = monte_carlo_dice_probability(num_simulations, target_sum_to_find)

print(f"Cílový součet: {target_sum_to_find}")
print(f"Odhadovaná pravděpodobnost: {estimated_prob:.4f}")
```
